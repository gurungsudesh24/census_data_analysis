```python
# encoding with label encoder only 
le = LabelEncoder()

for col in sample_df:
    sample_df[col] = le.fit_transform(sample_df[col])
sample_df
```

```python
#scaling the X
# min_max_scaler = MinMaxScaler()
# scaled_h_df = min_max_scaler.fit_transform(sample_df)
# print(scaled_h_df)
# scaled_df = pd.DataFrame(data = scaled_h_df, index=sample_df.index, columns= sample_df.columns)
# scaled_df

scaler = StandardScaler()
scaled_h_df = scaler.fit_transform(sample_df)
scaled_df = pd.DataFrame(data = scaled_h_df, index=sample_df.index, columns= sample_df.columns)
scaled_df
```

```python
# setting distance_threshold=0 ensures we compute the full tree.
cluster = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
cluster.fit(scaled_df)
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(cluster)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

### Transforming multidimensionality to one 

```python
#transform n variiables to 2 principal components to plot
pca = PCA(n_components=2)
pca_fit = pca.fit(scaled_df_h2)
principalComponents = pca_fit.transform(scaled_df_h2)
principalDf = pd.DataFrame(data = principalComponents
         , columns = ['principal component 1', 'principal component 2'])
```

```python
dataPointColor=[]
for row in h_cluster2.labels_:
    if row == 0: 
        dataPointColor.append('blue')
    else:
        dataPointColor.append('green')
                          
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'], 
c=dataPointColor, s=50, alpha=0.5)
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.show()
```
